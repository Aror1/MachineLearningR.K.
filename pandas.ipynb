9. What is the name with most occurrences?

```python
baby_names[["Name","Count"]].groupby("Name").sum().sort_values("Count",ascending = 0).head(1)
```

10. How many different names have the least occurrences?

```python
name_counts = baby_names.groupby('Name').size().reset_index(name='Count')
min_count = name_counts['Count'].min()
name_counts[name_counts['Count'] == min_count]
```

```python
len(name_counts[name_counts['Count'] == min_count])
```

11. What is the median name occurrence?

```python
name_counts = baby_names.groupby('Name').size().reset_index(name='Count')
median_occurrence = name_counts['Count'].median()
name_counts[(name_counts['Count'] >= median_occurrence - 1) & 
                           (name_counts['Count'] <= median_occurrence + 1)]
```
